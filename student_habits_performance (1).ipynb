### Data visualization

#### Correlation between study hours per day and exam score

```python
sns.scatterplot(data=df,x='study_hours_per_day',y='exam_score')
plt.title("study hours vs exam score")
plt.show()
          
```

#### Average exam score by diet quality

```python
plt.figure(figsize=(6,7))
diet_score = df.groupby('diet_quality')['exam_score'].mean().reset_index()
sns.barplot(data=diet_score,x='diet_quality',y='exam_score',color='violet',edgecolor='black')
plt.title('Average exam score by diet quality')
plt.show()
```

#### Distribution of study hours per day

```python
plt.hist(df['study_hours_per_day'],bins=10,color='blue',edgecolor='black')
plt.title('Distribution of study hours per day')
plt.xlabel('study hours per day')
plt.ylabel('Number of students')
plt.show()
```

#### Gender Distribution

```python
y=df['gender'].value_counts()
print(y)
plt.bar(y.index,y.values,width=0.3,color=('r','b'),edgecolor='black')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('count')
plt.show()
```

```python
avg_hours_by_gender.plot(kind='barh',color=['red','lightpink','violet'])
plt.ylabel("Average social media hours")
plt.title("Average time spent on social media by gender")
plt.show()
```

#### Does more sleep always improves mental health

```python
df.groupby('sleep_hours')['mental_health_rating'].mean().plot(kind='line')
plt.title("sleep hours vs Mental health")
```

#### Mental Health by part time job using a strip plot

```python
sns.stripplot(data=df,x='part_time_job',y='mental_health_rating',color='green')
plt.title('Mental health rating by part time job status')
plt.show()
```

#### Part time job status distribution

```python
job_counts = df['part_time_job'].value_counts()
plt.figure(figsize=(4,4))
plt.pie(job_counts,labels=job_counts.index,autopct='%1.1f%%',startangle=90,wedgeprops={'width':1.0})
plt.title("Part time jobs status distribution")
plt.tight_layout()
plt.show()
```

#### Mental health vs age

```python
age_health = df.groupby('age')['mental_health_rating'].mean().reset_index()
plt.figure(figsize=(4,4))
sns.lineplot(data=age_health,x='age',y='mental_health_rating',marker='o',color='red')
plt.title('Average mental health rating by age')
plt.tight_layout()
plt.show()
```
